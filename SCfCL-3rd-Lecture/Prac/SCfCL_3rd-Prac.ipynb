# コンピュータリテラシ発展 第3回 課題

今回の課題は**リスト型**，**タップル型**，**辞書型**，**for文**，**while文**に関する問題をそれぞれ3問ずつ合計15問となっています．

問題の最後には結果を表示してください．

## リスト（List）に関する課題

**問題1:** 空のリストを作成し，そのリストに整数1,2,3を追加してください．

**問題2:** リスト`[10,20,30,40,50]`の要素を逆順に並び替えてください．

**問題3:** リスト`[2,4,6,8,10]`の要素をすべて2倍にしてください．

## タプル（Tuple）に関する課題

**問題4:** 複数の整数を含むタプル`(1,2,3,4,5)`を作成し，その要素数を表示してください．

**問題5:** 問題4で作成したタプルを利用して，インデックス3の位置にある要素を表示してください．

**問題6:** タプル`(2,4,,6,8,10)`の要素を逆順に並び替えてください．

## 辞書（Dict）に関する課題

**問題7:** 空の辞書を作成し，キーが`name`で値（バリュー）が`John`の要素を追加してください．

**問題8:** 辞書 `{'name': 'Alice', 'age': 30, 'city': 'New York'}` からキー `'age'` に対応する値を取得してください．

**問題9:** 辞書 `{'x': 10, 'y': 20, 'z': 30}` のキーと値をすべて表示してください。

## 繰り返し（for文）に関する課題

**問題10:** 1から10までの整数をforループを使用して表示してください．

**問題11:** 1から100までの整数の合計をforループを使用して計算してください．

**問題12:** リスト `['cat', 'dog', 'rabbit', 'mouse']`の各要素をインデックスと共に表示してください．

## 繰り返し（while文）に関する課題

**問題13:** 1から20までの偶数をwhileループを使用して表示してください．

**問題14:** 1から50までの整数をwhileループを表示し，5と7の最小公倍数が表示されたら表示を終了してください．

**問題15:** 1月から12月までの英語表記のリストを作成して，whileループを表示し，今月(4月)が表示されたら表示を終了してください．
